```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

# Define the Lorenz system
def lorenz(t, state, sigma, beta, rho):
    x, y, z = state
    dxdt = sigma * (y - x)
    dydt = x * (rho - z) - y
    dzdt = x * y - beta * z
    return [dxdt, dydt, dzdt]

# Parameters
sigma = 10.0
beta = 8.0 / 3.0
rho = 28.0

# Initial conditions
initial_state = [1.0, 1.0, 1.0]

# Time vector
t_span = (0, 50)
t = np.linspace(t_span[0], t_span[1], 10000)

# Solve the system
solution = solve_ivp(
    lorenz,
    t_span,
    initial_state,
    args=(sigma, beta, rho),
    t_eval=t,
    method="RK45"
)

# Extract solutions
x, y, z = solution.y

# Plot the Lorenz attractor
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection="3d")
ax.plot(x, y, z, lw=0.5)
ax.set_title("Lorenz Attractor")
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
ax.set_zlabel("Z-axis")
plt.show()
```

```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

# Define the Lorenz system with noise
def lorenz_with_noise(t, state, sigma, beta, rho, noise_std):
    x, y, z = state
    dxdt = sigma * (y - x) + np.random.normal(0, noise_std)
    dydt = x * (rho - z) - y + np.random.normal(0, noise_std)
    dzdt = x * y - beta * z + np.random.normal(0, noise_std)
    return [dxdt, dydt, dzdt]

# Parameters
sigma = 10.0
beta = 8.0 / 3.0
rho = 28.0
noise_std = 0.5  # Standard deviation of noise

# Initial conditions
initial_state = [1.0, 1.0, 1.0]

# Time vector
t_span = (0, 50)
t = np.linspace(t_span[0], t_span[1], 10000)

# Solve the system
solution = solve_ivp(
    lorenz_with_noise,
    t_span,
    initial_state,
    args=(sigma, beta, rho, noise_std),
    t_eval=t,
    method="RK45"
)

# Extract solutions
x, y, z = solution.y

# Plot the noisy Lorenz attractor
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection="3d")
ax.plot(x, y, z, lw=0.5)
ax.set_title("Noisy Lorenz Attractor")
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
ax.set_zlabel("Z-axis")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# Parameters
A, B, C, D, E = 0.5, 0.1, 0.05, 0.01, 0.2  # Model coefficients
omega = 0.2  # Frequency of wave forcing
sigma_noise = 0.02  # Noise strength
grid_size = 20  # Spatial grid dimension (NxN)
t_start, t_end, num_time_points = 0, 200, 1000
t_eval = np.linspace(t_start, t_end, num_time_points)  # Time points for simulation

# Initialize state variables on a grid
U_init = np.random.uniform(9, 11, (grid_size, grid_size))  # Initial zonal wind
psi_init = np.random.uniform(0.9, 1.1, (grid_size, grid_size))  # Initial wave amplitude
T_init = np.random.uniform(0.4, 0.6, (grid_size, grid_size))  # Initial temperature

# Stochastic noise generator
def generate_noise(grid_size, scale=1.0):
    return scale * np.random.normal(size=(grid_size, grid_size))

# Define the multidimensional Holton-Mass model
def holton_mass_multidim(t, state):
    state = state.reshape(3, grid_size, grid_size)  # Reshape state to (3, NxN grid)
    U, psi, T = state

    # Add stochastic noise
    noise_U = generate_noise(grid_size, sigma_noise)
    noise_psi = generate_noise(grid_size, sigma_noise)
    noise_T = generate_noise(grid_size, sigma_noise)

    # Compute derivatives
    dU_dt = -B * U + C * psi - A * np.sin(omega * t) + noise_U
    dpsi_dt = -D * psi + T + noise_psi
    dT_dt = -E * T - psi * U + noise_T

    # Flatten the derivatives for the solver
    return np.array([dU_dt, dpsi_dt, dT_dt]).flatten()

# Initial condition (flattened)
initial_conditions = np.array([U_init, psi_init, T_init]).flatten()

# Solve the system
solution = solve_ivp(
    holton_mass_multidim, 
    (t_start, t_end), 
    initial_conditions, 
    t_eval=t_eval, 
    method="RK45"
)

# Extract results
time = solution.t
U, psi, T = solution.y.reshape(3, grid_size, grid_size, len(time))

# Visualize results for a single grid point
plt.figure(figsize=(12, 6))

# Choose a representative grid point
i, j = grid_size // 2, grid_size // 2

plt.subplot(3, 1, 1)
plt.plot(time, U[i, j, :], label=f"U at ({i}, {j})", color="blue")
plt.axhline(0, color="black", linestyle="--", linewidth=0.8)
plt.ylabel("Zonal Wind (U)")
plt.legend()
plt.grid()

plt.subplot(3, 1, 2)
plt.plot(time, psi[i, j, :], label=f"psi at ({i}, {j})", color="orange")
plt.axhline(0, color="black", linestyle="--", linewidth=0.8)
plt.ylabel("Wave Amplitude (psi)")
plt.legend()
plt.grid()

plt.subplot(3, 1, 3)
plt.plot(time, T[i, j, :], label=f"T at ({i}, {j})", color="green")
plt.axhline(0, color="black", linestyle="--", linewidth=0.8)
plt.ylabel("Temperature Perturbation (T)")
plt.xlabel("Time")
plt.legend()
plt.grid()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# Constants
num_layers = 27  # Vertical layers
H = 7e3  # Scale height in meters
dz = H / num_layers  # Vertical spacing
z = np.linspace(dz / 2, H - dz / 2, num_layers)  # Midpoint of each layer
A, B, C, D, E = 0.5, 0.1, 0.05, 0.02, 0.1  # Parameters
omega = 0.2  # Wave forcing frequency
sigma_noise = 0.02  # Noise strength

# State vector size
state_size = 3 * num_layers  # [Re(ψ), Im(ψ), U]

# Define the Holton-Mass model dynamics
def holton_mass_vertical(t, X):
    # Unpack state variables
    Re_psi = X[:num_layers]
    Im_psi = X[num_layers:2 * num_layers]
    U = X[2 * num_layers:]

    # Compute derivatives
    dU_dt = -B * U + C * Re_psi - A * np.sin(omega * t)  # Zonal wind
    dRe_psi_dt = -D * Re_psi + E * U  # Real part of wave amplitude
    dIm_psi_dt = -D * Im_psi  # Imaginary part of wave amplitude

    # Add stochastic noise
    dU_dt += sigma_noise * np.random.normal(size=U.shape)
    dRe_psi_dt += sigma_noise * np.random.normal(size=Re_psi.shape)
    dIm_psi_dt += sigma_noise * np.random.normal(size=Im_psi.shape)

    # Concatenate into a single derivative vector
    dX_dt = np.concatenate([dRe_psi_dt, dIm_psi_dt, dU_dt])
    return dX_dt

# Initial conditions
Re_psi_init = np.random.uniform(-0.1, 0.1, num_layers)  # Initial Re(ψ)
Im_psi_init = np.random.uniform(-0.1, 0.1, num_layers)  # Initial Im(ψ)
U_init = np.random.uniform(9, 11, num_layers)  # Initial U
X0 = np.concatenate([Re_psi_init, Im_psi_init, U_init])

# Time parameters
t_start, t_end, num_time_points = 0, 2000, 10000
t_eval = np.linspace(t_start, t_end, num_time_points)

# Solve the system
solution = solve_ivp(
    holton_mass_vertical,
    (t_start, t_end),
    X0,
    t_eval=t_eval,
    method="RK45"
)

# Extract solution
time = solution.t
Re_psi, Im_psi, U = np.split(solution.y, 3)  # Split solution into components

# Visualization
plt.figure(figsize=(12, 8))

# Plot U (zonal wind) for a few vertical layers
for i in range(0, num_layers, 5):  # Plot every 5th layer
    plt.plot(time, U[i, :], label=f"U at z={z[i]:.0f} m")

plt.title("Holton-Mass Model: Zonal Wind Evolution")
plt.xlabel("Time")
plt.ylabel("Zonal Wind (U)")
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the Lorenz system
sigma = 10.0
beta = 8 / 3
rho = 28.0

# Noise strength
noise_strength = 1

# Lorenz system equations with noise
def lorenz_sde(state, t):
    x, y, z = state
    dxdt = sigma * (y - x)
    dydt = x * (rho - z) - y
    dzdt = x * y - beta * z
    return np.array([dxdt, dydt, dzdt])

# Euler-Maruyama method for SDEs
def euler_maruyama(f, state0, t, noise_strength):
    n_steps = len(t)
    dt = t[1] - t[0]
    n_dim = len(state0)
    trajectory = np.zeros((n_steps, n_dim))
    trajectory[0] = state0
    
    for i in range(1, n_steps):
        dw = np.sqrt(dt) * np.random.normal(size=n_dim)  # Wiener increment
        trajectory[i] = trajectory[i-1] + f(trajectory[i-1], t[i-1]) * dt + noise_strength * dw
    
    return trajectory

# Initial condition and time span
state0 = [1.0, 1.0, 1.0]
t = np.linspace(0, 50, 10000)

np.random.seed(0)
# Solve the SDE
trajectory = euler_maruyama(lorenz_sde, state0, t, noise_strength)

# Extract components
x, y, z = trajectory.T

# Plot the results
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot(x, y, z, color="blue", linewidth=0.5)
ax.set_title("Lorenz System with Noise (SDE Integration)")
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
plt.show()
```
